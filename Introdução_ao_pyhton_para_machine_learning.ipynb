```python
dataFrame.dropna() -> Se quiserem remover linhas com valores em falta ou não importante

dataFrame["Idade"].fillna(dataFrame["Idade"].mean(), inplace=True) # 25, 30, NaN, 40 || 25+30+40/3 = 30


"""
nome      idade
joao       25
bruno      30
joana      NaN - 30 (a media de 20, 30, 40)
maria      40


"""
dataFrame["Cidade"].fillna("Desconhecido", inplace=True) # usar em colunas ctaegorica spar apreencher com texto padrão

dataFrame.fillna(0) # Preencher com valores fixos somentes em testes ou dataSets simples

dataFrame.duplicate().sum() # soma a quantidade de linhas que estão duplicadas
dataFrame.drop_duplicates(inplace=True)

#NORMALIZAÇÃO DOS DADOS DE CATEGORIAS
dataFrame["Comprou"] = dataFrame["Comprou"].str.lower().str.strip()
dataFrame["Comprou"] = dataFrame["Comprou"].replace({"y":"sim", "n":"não", "s": "sim"})

dataFrame.dtypes # verificar os tipos de dados
dataFrame["Idade"] = pd.to_numeric(dataFrame["Idade"], errors = "coerce")

# CONVERTER COLUNAS CATEGORIAS EM NUMEROS (PREPARAÇÃO PARA ML)
dataFrame["Comprou"] = dataFrame["Comprou"].map({"sim" : 1, "não": 0})


dataFrame["Comprou"] = dataFrame["Comprou"].map(["SIM", "Sim", "S", "Y", "y"], 1)

```
